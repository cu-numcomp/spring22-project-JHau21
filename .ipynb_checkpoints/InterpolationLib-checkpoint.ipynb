### Library: Interpolation.jl

For this project I decided to explore a library called Interpolation.jl that essentially maintains a series of different interpolation methods, mainly intended for use on a set of discrete data points. 

The library is designed to be easy to use, provide support for a broad range of algorithms, and be extremely efficient and quick. Currently the best support is for B-Spline interpolation, but the library is intended and being built to support a much broader range of methods. For B-Spline the package supports 4 degrees.
- Contant
- Linear
- Quadratic
- Cubic

In terms of speed and efficiency, the package uses two different benchmarks to show off performance.

- Construction Time - the time is takes method to initialize an interpolation object. In this case, faster is better because ideally the object being used can quickly initialize everything and be prepared to perform actual interpolation. In essence, we want the object to be constructed extremely quickly to improve it's performance.
    
- Throughput - the number of interpolations that can be done per second. In this case, larger values are better because performing more interpolations per second indicates faster construction of an interpolating function or model.
    
In order to present these interpolation methods and objects being faster and more efficient, Interpolations.jl compared their methods and objects with the those of other Julia libraries that similarly allow for interpolation and extrapolation. To see the actual comparison scroll down to the **Performance Shootout** of https://github.com/JuliaMath/Interpolations.jl/blob/master/README.md. 

I would now to show off some of the libraries abilites using their interpolation objects and graphing everything.

Let's start by importing the library for use.

```julia
using Interpolations
```

### The Basics

To begin we can start by understanding how to actually create an interpolation object and understand how these objects function.

#### What is an Interpolation Object?

According to the project managers it is essentially an array that supports real number evaluation.

To me, this meant that interpolation objects are pretty much just a series of points that are intended to be evaluated via functions using real numbers. 

#### How can Interpolation Objects be created for use?

In order to create an interpolation object, the user can simply type `interpolate` to call the interpolate constructor, assuming the package is being in the file.

`interpolate` then takes a series of parameters, the data structure containing all of the data to be interpolated, and the interpolation configurations.

Let's create a small test interpolation object.

```julia
xArr = 1:1:10
test = interpolate(xArr, BSpline(Linear()))
```

#### What now?

In the above code, we created a simply array of 10 values, equally spaced by 1.

We then created the interpolation object and specified BSpline-Linear interpolation for the configurations.

How does it look? First let's import Plots and then graph everything.

```julia
using Plots
```

```julia
f(x) = x # function to map x inputs to y outputs

yArr = f.(xArr) # create y output array for graphing scatter of points

yLin = [test(x) for x in xArr] # create function to apply the interpolation to the values in xArr

scatter(yArr, label = "Data Points") # scatter the points
plot!(yLin, label = "Interp. Function") # plot everything
plot!(size = (800, 500))
```

#### Analysis

In our above example, we did a couple things.

First, we created a simply array of 10 values from 1 to 10, then declared our interpolate object, and specific BSpline-Linear interpolation.

Next, we created a simple function to map our x-values to corresponding y-values, and stored those results in another array.

Then, we passed each of our x-values to our interpolate object for interpolation and stored the outputs in an array.

After scattering our values then plotting our interpolation object, we can see a very successful and accurate interpolation on a fairly basic and obviously linear dataset. So, let's try something more complicated.

What about passing the same set of points to a sin function instead to create a non-linear fit.

```julia
x = 1:1:10 # same set of points 1 through 10, equally spaced

f(x) = sin(x) # new function to create more complex datapoints by running everything through a sin function

y = f.(x) # create set of data points

nonLinInt = interpolate(y, BSpline(Linear())) # same interpolation configuration and points

yLin = [nonLinInt(x) for x in x] # create function to apply the interpolation to the values in xArr

scatter(y, label = "Data Points") # scatter the points
plot!(yLin, label = "Interp. Function") # plot everything
plot!(size = (800, 500))
```

#### Analysis and Improvement

In the above example, we did essentially the exact same thing again, but instead of a set of linear data points, we created a set of sinusoidal points. As we can see from the figure above, this produces a fairly unsatisfying results. Lukcily, we have a variety of other Interpolation methods to choose from.
